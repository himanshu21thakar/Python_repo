```
# âž• Python Math & Logic Operations - Detailed with Examples

# --- Arithmetic Operators ---
# These operators are used to perform mathematical operations

a = 10
b = 3

print("Addition (a + b):", a + b)
print("Subtraction (a - b):", a - b)
print("Multiplication (a * b):", a * b)
print("Division (a / b):", a / b)  # Floating-point division
print("Floor Division (a // b):", a // b)  # Discards remainder
print("Modulus (a % b):", a % b)  # Remainder of division
print("Exponentiation (a ** b):", a ** b)  # a to the power b

# --- Comparison Operators ---
# These return True or False depending on the comparison

print("Is a equal to b?", a == b)
print("Is a not equal to b?", a != b)
print("Is a greater than b?", a > b)
print("Is a less than or equal to b?", a <= b)

# --- Logical Operators ---
# These are used to combine conditional statements

x = True
y = False

print("x and y:", x and y)  # True only if both are True
print("x or y:", x or y)    # True if at least one is True
print("not x:", not x)      # Reverses the boolean value

# --- Assignment Operators ---
# These assign values to variables with operations

c = 5
c += 2  # Same as c = c + 2
print("After += operation:", c)

c *= 3  # Same as c = c * 3
print("After *= operation:", c)

# --- Identity Operators ---
# Used to compare objects, not just values

list1 = [1, 2]
list2 = list1
list3 = [1, 2]

print("list1 is list2:", list1 is list2)  # True (same memory)
print("list1 is list3:", list1 is list3)  # False (different memory)
print("list1 == list3:", list1 == list3)  # True (values are same)

# --- Membership Operators ---
# Used to test if a value is present in a sequence

nums = [1, 2, 3, 4]
print("Is 3 in nums?", 3 in nums)
print("Is 5 not in nums?", 5 not in nums)

# --- Math Module ---
# The math module provides many mathematical functions

import math

print("Square root of 16:", math.sqrt(16))
print("Ceiling of 3.2:", math.ceil(3.2))
print("Floor of 3.7:", math.floor(3.7))
print("Value of pi:", math.pi)
print("Sine of 90 degrees (in radians):", math.sin(math.radians(90)))

# --- Practice Task ---
# Try calculating the area of a circle with radius 7 using math.pi
radius = 7
area = math.pi * (radius ** 2)
print("Area of circle with radius 7:", area)
```
